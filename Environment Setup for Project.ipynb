# Environment Setup for the retail project
### Creating External location for the retaildata container

```sql
    
create external location retail_ext
url 'abfss://retaildata@dbstorageproject2.dfs.core.windows.net/'
with (storage credential  `cred_stg_retail`)
comment "my external location"
```

## Create Catalog

```sql
create catalog retails
managed location 'abfss://retaildata@dbstorageproject2.dfs.core.windows.net/'
```

## Create Schema 

```sql
create schema if not exists retails.bronze
managed location 'abfss://retaildata@dbstorageproject2.dfs.core.windows.net/bronze';

create schema if not exists retails.silver
managed location 'abfss://retaildata@dbstorageproject2.dfs.core.windows.net/silver';

create schema if not exists retails.gold
managed location 'abfss://retaildata@dbstorageproject2.dfs.core.windows.net/gold';

create schema if not exists retails.landing
managed location 'abfss://retaildata@dbstorageproject2.dfs.core.windows.net/landing';
```

### Creating External Volume for data accessing

```sql
use catalog retails;
use schema landing;

create external volume if not exists operational_data
location 'abfss://retaildata@dbstorageproject2.dfs.core.windows.net/landing/operational_data/'
comment 'My volume'
```
