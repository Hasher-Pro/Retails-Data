### Now we are having issue while reading the file because it is showing csv in one column so we are reading it through read_files

```sql
create or replace table retails.bronze.orders as
select * from read_files('/Volumes/retails/landing/operational_data/',
format => 'csv',
delimiter => ',',
header => 'true')
```

```sql
select * from retails.bronze.orders_casted
```

#### we are having issues in the above data (
  - orders date should be date and not string.
)

```sql
CREATE OR REPLACE TABLE retails.bronze.orders_casted AS
SELECT 
    order_id,
    customer_id,
    product_id,

    COALESCE(
        try_to_date(order_date, 'yyyy-MM-dd'),
        try_to_date(order_date, 'MM/dd/yyyy'),
         try_to_date(order_date, 'M/d/yyyy')
    ) AS order_date,

    quantity,
    price,
    payment_method,
    order_status

FROM retails.bronze.orders;
```
