### Ingestion of orders data in bronze using structured streaming

```sql
create or refresh streaming table retails.bronze.orders_struct_stream
tblproperties ('quality'='bronze')
as 
select *,
_metadata.file_path as file_path,
current_timestamp as ingestion_time
from stream read_files('/Volumes/retails/landing/operational_data/orders_*', format => 'csv')
```
